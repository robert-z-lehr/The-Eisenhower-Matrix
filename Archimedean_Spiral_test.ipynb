```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parametric functions
def x(t):
    return np.cos(t)

def y(t):
    return np.sin(t)

def z(t):
    return t

# Generate a range of t values
t_values = np.linspace(0, 10, 1000)

# Generate the x, y, z values
x_values = x(t_values)
y_values = y(t_values)
z_values = z(t_values)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x_values, y_values, z_values)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# Import Dependencies
import turtle

def hilbert_curve(turtle, iterations, angle=90, length=200.0):
    if iterations == 0:
        return
    turtle.left(angle)
    hilbert_curve(turtle, iterations-1, -angle, length/2)
    turtle.forward(length)
    turtle.right(angle)
    hilbert_curve(turtle, iterations-1, angle, length/2)
    turtle.forward(length)
    hilbert_curve(turtle, iterations-1, angle, length/2)
    turtle.right(angle)
    turtle.forward(length)
    hilbert_curve(turtle, iterations-1, -angle, length/2)
    turtle.left(angle)

# Setup Turtle graphics
window = turtle.Screen()
window.bgcolor("white")

hilbert = turtle.Turtle()
hilbert.speed(0)  # Fastest speed

# Draw Hilbert Curve
hilbert.penup()
hilbert.goto(-50, -50)
hilbert.pendown()
hilbert_curve(hilbert, 4, length=100)

# Exit on click
window.exitonclick()
```
